```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset from the provided paths
class1 = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\rd_group4\class1.txt', delimiter=' ', header=None)
class2 = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\rd_group4\class2.txt', delimiter=' ', header=None)
class3 = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\rd_group4\class3.txt', delimiter=' ', header=None)

# Dynamically create column names based on the number of columns
for dataset in [class1, class2, class3]:
    num_columns = dataset.shape[1]
    column_names = ['X', 'Y'] + [f'Feature_{i}' for i in range(3, num_columns + 1)]
    dataset.columns = column_names

# Combine the datasets
combined_dataset = pd.concat([class1, class2, class3], ignore_index=True)

# Assign class labels
class_labels = np.concatenate([np.zeros(len(class1)), np.ones(len(class2)), np.full(len(class3), 2)])
combined_dataset['Class'] = class_labels

# Features and labels
X = combined_dataset.iloc[:, :2].to_numpy()  # Use only the first two columns for features (X, Y) to allow plotting
y = combined_dataset['Class'].to_numpy()     # Last column is the class label

# Split the data into 70% training and 30% testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Helper functions for GDA
def calculate_mean_covariance(X_train, y_train):
    classes = np.unique(y_train)
    mean_vectors = {}
    covariance_matrices = {}
    for cls in classes:
        X_cls = X_train[y_train == cls]
        mean_vectors[cls] = np.mean(X_cls, axis=0)
        covariance_matrices[cls] = np.cov(X_cls, rowvar=False)
    return mean_vectors, covariance_matrices

def discriminant_function(x, mu, sigma, prior):
    sigma_inv = np.linalg.inv(sigma)
    det_sigma = np.linalg.det(sigma)
    mahalanobis_term = -0.5 * np.dot(np.dot((x - mu).T, sigma_inv), (x - mu))
    log_det_term = -0.5 * np.log(det_sigma)
    log_prior_term = np.log(prior)
    return mahalanobis_term + log_det_term + log_prior_term

def classify(x, means, covariances, priors):
    g_values = []
    for cls in means.keys():
        g_value = discriminant_function(x, means[cls], covariances[cls], priors[cls])
        g_values.append(g_value)
    return np.argmax(g_values)

# Calculate class-wise statistics (means and covariances)
mean_vectors, covariance_matrices = calculate_mean_covariance(X_train, y_train)

# Display covariance matrices for each class
print("Covariance Matrices:")
for cls, cov_matrix in covariance_matrices.items():
    print(f"\nClass {int(cls)} Covariance Matrix:\n{cov_matrix}")

# Calculate priors
class_counts = {cls: np.sum(y_train == cls) for cls in np.unique(y_train)}
total_samples = len(y_train)
priors = {cls: class_counts[cls] / total_samples for cls in class_counts.keys()}

# Classify the test data
y_pred = np.array([classify(x, mean_vectors, covariance_matrices, priors) for x in X_test])

# Compute confusion matrix
def calculate_confusion_matrix(y_true, y_pred):
    unique_classes = np.unique(np.concatenate((y_true, y_pred)))
    n_classes = len(unique_classes)
    confusion_matrix = np.zeros((n_classes, n_classes), dtype=int)
    for i in range(len(y_true)):
        true_class = int(y_true[i])
        predicted_class = int(y_pred[i])
        confusion_matrix[true_class, predicted_class] += 1
    return confusion_matrix

# Plot decision boundary
def plot_decision_boundary(X, y, mean_vectors, covariance_matrices, priors):
    """
    Plots the decision boundaries of the GDA classifier.
    """
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    Z = np.array([classify(np.array([x, y]), mean_vectors, covariance_matrices, priors)
                  for x, y in zip(xx.ravel(), yy.ravel())])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
    plt.title("Decision Boundary")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()

# Call the visualization function
plot_decision_boundary(X_test, y_test, mean_vectors, covariance_matrices, priors)
```

```python
from scipy.stats import multivariate_normal

def plot_contours_and_decision_boundary(X_train, y_train, mean_vectors, covariance_matrices, priors):
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
    grid_points = np.c_[xx.ravel(), yy.ravel()]

    # Evaluate decision boundary
    Z = np.array([classify(np.array([x, y]), mean_vectors, covariance_matrices, priors) for x, y in grid_points])
    Z = Z.reshape(xx.shape)

    # Plot contour for each class
    plt.figure(figsize=(10, 8))
    for cls, mean in mean_vectors.items():
        cov = covariance_matrices[cls]
        rv = multivariate_normal(mean=mean, cov=cov)
        pdf_values = rv.pdf(grid_points).reshape(xx.shape)
        plt.contour(xx, yy, pdf_values, levels=5, alpha=0.7, colors='black')

    # Plot decision boundary
    plt.contourf(xx, yy, Z, alpha=1, cmap=plt.cm.coolwarm)

    # Scatter plot of training data
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolor='k', cmap=plt.cm.coolwarm, s=50)

    plt.title("Contour Plots and Decision Boundary")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.colorbar()
    plt.show()

plot_contours_and_decision_boundary(X_train, y_train, mean_vectors, covariance_matrices, priors)
```

```python
def calculate_mean_covariance_isotropic(X_train, y_train):
    """
    Calculates the mean vector and shared isotropic covariance matrix for all classes.
    """
    classes = np.unique(y_train)
    mean_vectors = {}
    all_data = []

    for cls in classes:
        X_cls = X_train[y_train == cls]
        mean_vectors[cls] = np.mean(X_cls, axis=0)
        all_data.append(X_cls)

    # Combine all class data to compute shared isotropic covariance
    all_data = np.vstack(all_data)
    variances = np.var(all_data, axis=0)
    isotropic_variance = np.mean(variances)  # Average variance
    isotropic_covariance_matrix = np.eye(all_data.shape[1]) * isotropic_variance

    covariance_matrices = {cls: isotropic_covariance_matrix for cls in classes}

    return mean_vectors, covariance_matrices

def classify(x, means, covariances, priors):
    """
    Classifies a single data point based on discriminant function values.
    """
    g_values = []
    for cls in means.keys():
        sigma_inv = np.linalg.inv(covariances[cls])
        det_sigma = np.linalg.det(covariances[cls])
        d = len(x)

        mahalanobis_term = -0.5 * np.dot(np.dot((x - means[cls]).T, sigma_inv), (x - means[cls]))
        log_det_term = -0.5 * np.log(det_sigma)
        log_prior_term = np.log(priors[cls])

        g_value = mahalanobis_term + log_det_term + log_prior_term
        g_values.append(g_value)

    return np.argmax(g_values)

def plot_decision_boundary_isotropic(X, y, mean_vectors, covariance_matrices, priors):
    """
    Plots the decision boundaries of the modified GDA classifier with isotropic covariance matrix.
    """
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    Z = np.array([classify(np.array([x, y]), mean_vectors, covariance_matrices, priors)
                  for x, y in zip(xx.ravel(), yy.ravel())])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
    plt.title("Decision Boundaries with Isotropic Covariance Matrix")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()

# Calculate class-wise statistics (means and shared isotropic covariance)
mean_vectors_isotropic, covariance_matrices_isotropic = calculate_mean_covariance_isotropic(X_train, y_train)

# Print the isotropic covariance matrix
print("Shared Isotropic Covariance Matrix:")
print(next(iter(covariance_matrices_isotropic.values())))

# Calculate priors
class_counts = {cls: np.sum(y_train == cls) for cls in np.unique(y_train)}
total_samples = len(y_train)
priors = {cls: class_counts[cls] / total_samples for cls in class_counts.keys()}

# Plot the decision boundary with isotropic covariance matrix
plot_decision_boundary_isotropic(X_test, y_test, mean_vectors_isotropic, covariance_matrices_isotropic, priors)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

def calculate_mean_covariance_shared(X_train, y_train):
    classes = np.unique(y_train)
    mean_vectors = {}
    all_data = []

    for cls in classes:
        X_cls = X_train[y_train == cls]
        mean_vectors[cls] = np.mean(X_cls, axis=0)
        all_data.append(X_cls)

    # Combine all class data to compute shared covariance matrix
    all_data = np.vstack(all_data)
    shared_covariance_matrix = np.cov(all_data, rowvar=False)  # Full covariance matrix

    covariance_matrices = {cls: shared_covariance_matrix for cls in classes}

    return mean_vectors, covariance_matrices

def classify(x, means, covariances, priors):
    g_values = []
    for cls in means.keys():
        sigma_inv = np.linalg.inv(covariances[cls])
        det_sigma = np.linalg.det(covariances[cls])
        d = len(x)

        mahalanobis_term = -0.5 * np.dot(np.dot((x - means[cls]).T, sigma_inv), (x - means[cls]))
        log_det_term = -0.5 * np.log(det_sigma)
        log_prior_term = np.log(priors[cls])

        g_value = mahalanobis_term + log_det_term + log_prior_term
        g_values.append(g_value)

    return np.argmax(g_values)

def plot_decision_boundary_shared(X, y, mean_vectors, covariance_matrices, priors):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))
    grid_points = np.c_[xx.ravel(), yy.ravel()]

    Z = np.array([classify(np.array([x, y]), mean_vectors, covariance_matrices, priors)
                  for x, y in zip(xx.ravel(), yy.ravel())])
    Z = Z.reshape(xx.shape)

    # Create filled contour plot for class predictions
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)

    # Plot class-wise contours based on the Gaussian PDF
    for cls, mean in mean_vectors.items():
        cov = covariance_matrices[cls]
        rv = multivariate_normal(mean=mean, cov=cov)
        pdf_values = rv.pdf(grid_points).reshape(xx.shape)
        plt.contour(xx, yy, pdf_values, levels=5, alpha=0.7, colors='black')

    # Scatter plot of the data points
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
    
    plt.title("Decision Boundaries with Shared Covariance Matrix and Class-wise Contours")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()

# Assuming X_train, y_train, X_test, y_test are already defined

# Calculate class-wise statistics (means and shared covariance)
mean_vectors_shared, covariance_matrices_shared = calculate_mean_covariance_shared(X_train, y_train)

# Print the shared covariance matrix
print("Shared Covariance Matrix:")
print(next(iter(covariance_matrices_shared.values())))

# Calculate priors
class_counts = {cls: np.sum(y_train == cls) for cls in np.unique(y_train)}
total_samples = len(y_train)
priors = {cls: class_counts[cls] / total_samples for cls in class_counts.keys()}

# Plot the decision boundary with shared covariance matrix
plot_decision_boundary_shared(X_test, y_test, mean_vectors_shared, covariance_matrices_shared, priors)
```
