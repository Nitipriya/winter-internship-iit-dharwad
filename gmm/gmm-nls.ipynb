```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\NLS_Group04.txt', delimiter=' ', header=None)

# Dynamically create column names based on the number of columns
num_columns = data.shape[1]
column_names = ['X', 'Y'] + [f'Feature_{i}' for i in range(3, num_columns + 1)]
data.columns = column_names

# Assign class labels (not used for clustering)
class_labels = np.concatenate([np.zeros(300), np.ones(500), np.full(1000, 2)])
data['Class'] = class_labels

# Select features for clustering
X = data[['X', 'Y']].values

# GMM Implementation
class GMM:
    def __init__(self, n_components, max_iter=100, tol=1e-3):
        self.n_components = n_components
        self.max_iter = max_iter
        self.tol = tol
        self.log_likelihoods = []  # Store log-likelihoods for plotting

    def initialize_parameters(self, X):
        """Initialize parameters using KMeans centroids."""
        kmeans = KMeans(n_clusters=self.n_components, random_state=42)
        kmeans.fit(X)
        self.means = kmeans.cluster_centers_  # Initialize means
        self.covariances = [np.cov(X.T) for _ in range(self.n_components)]  # Initialize covariances
        self.weights = np.full(self.n_components, 1 / self.n_components)  # Equal weights

    def gaussian_pdf(self, X, mean, cov):
        """Calculate Gaussian probability density function."""
        d = X.shape[1]
        cov_inv = np.linalg.inv(cov)
        diff = X - mean
        exponent = np.exp(-0.5 * np.sum(diff @ cov_inv * diff, axis=1))
        denominator = np.sqrt(((2 * np.pi) ** d) * np.linalg.det(cov))
        return exponent / denominator

    def e_step(self, X):
        """Expectation step: calculate responsibilities."""
        self.resp = np.zeros((X.shape[0], self.n_components))
        for k in range(self.n_components):
            self.resp[:, k] = self.weights[k] * self.gaussian_pdf(X, self.means[k], self.covariances[k])
        self.resp /= self.resp.sum(axis=1, keepdims=True)  # Normalize responsibilities

    def m_step(self, X):
        """Maximization step: update parameters."""
        Nk = self.resp.sum(axis=0)
        self.weights = Nk / X.shape[0]
        self.means = (self.resp.T @ X) / Nk[:, np.newaxis]
        self.covariances = []
        for k in range(self.n_components):
            diff = X - self.means[k]
            cov = (self.resp[:, k, np.newaxis] * diff).T @ diff / Nk[k]
            self.covariances.append(cov)

    def compute_log_likelihood(self, X):
        """Compute log-likelihood."""
        likelihood = np.sum(
            np.log(np.sum([self.weights[k] * self.gaussian_pdf(X, self.means[k], self.covariances[k])
                           for k in range(self.n_components)], axis=0))
        )
        return likelihood

    def fit(self, X):
        """Train the GMM using the EM algorithm."""
        self.initialize_parameters(X)
        log_likelihood_prev = -np.inf
        
        for iteration in range(self.max_iter):
            self.e_step(X)
            self.m_step(X)
            
            # Calculate log-likelihood
            log_likelihood = self.compute_log_likelihood(X)
            self.log_likelihoods.append(log_likelihood)
            
            if np.abs(log_likelihood - log_likelihood_prev) < self.tol:
                break
            log_likelihood_prev = log_likelihood

    def predict(self, X):
        """Predict cluster assignments for each point."""
        self.e_step(X)
        return np.argmax(self.resp, axis=1)

# Initialize and fit the GMM
gmm = GMM(n_components=3, max_iter=100, tol=1e-3)
gmm.fit(X)
gmm_labels = gmm.predict(X)

# Plot GMM clustering results
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, cmap='viridis', s=20, alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('GMM Clustering Results ')
plt.colorbar(label='Cluster Label')
plt.show()

# Plot log-likelihood convergence
plt.figure(figsize=(10, 7))
plt.plot(range(1, len(gmm.log_likelihoods) + 1), gmm.log_likelihoods, marker='x', linestyle='-')
plt.xlabel('Iterations')
plt.ylabel('Log Likelihood')
plt.title('Log-Likelihood Convergence')
plt.show()

# Print parameters
print("GMM Means:\n", gmm.means)
print("GMM Covariances:\n", gmm.covariances)
print("GMM Weights:\n", gmm.weights)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the Mahalanobis distance function
def mahalanobis_dist(x1, x2, cov_matrix):
    mean_diff = x1 - x2  
    cov_matrix_inv = np.linalg.inv(cov_matrix)  
    distance = np.sqrt(np.dot(np.dot(mean_diff.T, cov_matrix_inv), mean_diff))
    return distance

# Define K-Means clustering function
def KMeansClustering(X, K, maxIteration, mod: bool, tolerance=1e-4):
    m = X.shape[0]
    np.random.seed(42)
    centroids = X[np.random.choice(m, K, replace=False)]
    idx = np.zeros(m, dtype=int)
    prevCentroid = np.zeros_like(centroids)
    cov = np.cov(X, rowvar=False)
    print("Covariance Matrix:\n", cov)
    print("\nMean of the data:\n", np.mean(X, axis=0))
    
    iteration_list = []
    centroid_movements = []

    for i in range(maxIteration):
        if mod:
            dist = np.zeros((m, K))
            for j in range(m):
                for k in range(K):
                    dist[j, k] = mahalanobis_dist(X[j], centroids[k], cov)
        else: 
            dist = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        idx = np.argmin(dist, axis=1)
        prevCentroid = centroids.copy()
        for k in range(K):
            points = X[idx == k]
            if len(points) > 0:
                centroids[k] = np.mean(points, axis=0)
        centroid_movement = np.linalg.norm(centroids - prevCentroid)
        iteration_list.append(i + 1)
        centroid_movements.append(centroid_movement)
        if centroid_movement < tolerance:
            print(f"Converged in {i + 1} iterations!")
            break

    # Plot centroid movement
    plt.figure(figsize=(8, 5))
    plt.plot(iteration_list, centroid_movements, marker='o', label="Centroid Movement")
    plt.title("Centroid Movement vs. Iterations")
    plt.xlabel("Iteration")
    plt.ylabel("Centroid Movement")
    plt.grid(True)
    plt.legend()
    plt.show()
        
    return centroids, idx

# Load the dataset
data = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\NLS_Group04.txt', delimiter=r'\s+', header=None)

# Dynamically generate column names based on actual data dimensions
num_columns = data.shape[1]
column_names = ['X', 'Y'] + [f'Feature_{i}' for i in range(3, num_columns + 1)]
data.columns = column_names

# Assign labels for three classes
class_labels = np.concatenate([np.zeros(300), np.ones(500), np.full(1000, 2)])
data['Class'] = class_labels

# Scatter plot for three classes
plt.figure(figsize=(10, 7))
plt.scatter(data[data['Class'] == 0]['X'], data[data['Class'] == 0]['Y'], color='red', label='Class 1', alpha=0.7)
plt.scatter(data[data['Class'] == 1]['X'], data[data['Class'] == 1]['Y'], color='blue', label='Class 2', alpha=0.7)
plt.scatter(data[data['Class'] == 2]['X'], data[data['Class'] == 2]['Y'], color='green', label='Class 3', alpha=0.7)

plt.title("Scatter Plot of Classes", fontsize=16)
plt.xlabel("X-axis", fontsize=14)
plt.ylabel("Y-axis", fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Extract features (X, Y) and labels
X = data[['X', 'Y']].values  # Features (X and Y)
y = data['Class'].values  # Labels (Class 1, 2, 3)

# Train-test split function
def train_test_split(data, test_size=0.3):
    np.random.shuffle(data)
    split_idx = int((1 - test_size) * len(data))
    return data[:split_idx], data[split_idx:]

X_train, X_test = train_test_split(X, test_size=0.3)

# Perform K-Means clustering with Euclidean distance (K=3 for three classes)
centroidEuclid, idxEuclid = KMeansClustering(X, K=3, maxIteration=100, mod=False)

# Perform K-Means clustering with Mahalanobis distance (K=3 for three classes)
centroidMahalanobis, idxMahalanobis = KMeansClustering(X, K=3, maxIteration=100, mod=True)

# Plot Euclidean results
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=idxEuclid, cmap='viridis', s=30, label='Data Points')
plt.scatter(centroidEuclid[:, 0], centroidEuclid[:, 1], color='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering (Euclidean Distance)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

# Plot Mahalanobis results
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=idxMahalanobis, cmap='viridis', s=30, label='Data Points')
plt.scatter(centroidMahalanobis[:, 0], centroidMahalanobis[:, 1], color='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering (Mahalanobis Distance)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
