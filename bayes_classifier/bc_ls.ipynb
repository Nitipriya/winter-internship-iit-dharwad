```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Load datasets with space delimiter
class1 = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\LS_Group04\Class1.txt', delimiter=' ', header=None)
class1.columns = ['X', 'Y']
class2 = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\LS_Group04\Class2.txt', delimiter=' ', header=None)
class2.columns = ['X', 'Y']
class3 = pd.read_csv(r'C:\Users\nitip\OneDrive\Desktop\dec2024\Group04[1]\Group04\LS_Group04\Class3.txt', delimiter=' ', header=None)
class3.columns = ['X', 'Y']
# Combine the datasets
combined_dataset = pd.concat([class1, class2, class3], ignore_index=True)

# Features and labels
X = combined_dataset[['X', 'Y']].to_numpy()  # Features (X and Y)
y = np.concatenate([np.zeros(len(class1)), np.ones(len(class2)), np.full(len(class3), 2)])  # Labels: Class 1, 2, 3

# Split the data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Helper functions for GDA
def calculate_mean_covariance(X_train, y_train):
    """
    Calculates the mean vector and covariance matrix for each class.
    """
    classes = np.unique(y_train)
    mean_vectors = {}
    covariance_matrices = {}

    for cls in classes:
        X_cls = X_train[y_train == cls]
        mean_vectors[cls] = np.mean(X_cls, axis=0)
        covariance_matrices[cls] = np.cov(X_cls, rowvar=False)
    
    return mean_vectors, covariance_matrices

def discriminant_function(x, mu, sigma, prior):
    """
    Computes the discriminant function value for a single class.
    """
    sigma_inv = np.linalg.inv(sigma)
    det_sigma = np.linalg.det(sigma)
    d = len(x)

    # Mahalanobis distance term
    mahalanobis_term = -0.5 * np.dot(np.dot((x - mu).T, sigma_inv), (x - mu))

    # Log determinant and prior terms
    log_det_term = -0.5 * np.log(det_sigma)
    log_prior_term = np.log(prior)

    return mahalanobis_term + log_det_term + log_prior_term

def classify(x, means, covariances, priors):
    """
    Classifies a single data point based on discriminant function values.
    """
    g_values = []
    for cls in means.keys():
        g_value = discriminant_function(x, means[cls], covariances[cls], priors[cls])
        g_values.append(g_value)
    return np.argmax(g_values)

# Calculate class-wise statistics (means and covariances)
mean_vectors, covariance_matrices = calculate_mean_covariance(X_train, y_train)

# Display covariance matrices for each class
print("Covariance Matrices:")
for cls, cov_matrix in covariance_matrices.items():
    print(f"\nClass {int(cls)} Covariance Matrix:\n{cov_matrix}")

# Calculate priors
class_counts = {}
for cls in np.unique(y_train):
    class_counts[cls] = np.sum(y_train == cls)
total_samples = len(y_train)
priors = {cls: class_counts[cls] / total_samples for cls in class_counts.keys()}

# Classify the test data
y_pred = np.array([classify(x, mean_vectors, covariance_matrices, priors) for x in X_test])

# Visualization of decision boundaries (optional)
def plot_decision_boundary(X, y, mean_vectors, covariance_matrices, priors):
    """
    Plots the decision boundaries of the GDA classifier.
    """
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    Z = np.array([classify(np.array([x, y]), mean_vectors, covariance_matrices, priors)
                  for x, y in zip(xx.ravel(), yy.ravel())])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
    plt.title("Decision Boundary")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()

# Call the visualization function
plot_decision_boundary(X_test, y_test, mean_vectors, covariance_matrices, priors)
```

```python

def calculate_confusion_matrix(y_true, y_pred):
    """
    Computes the confusion matrix manually.
    """
    unique_classes = np.unique(np.concatenate((y_true, y_pred)))
    n_classes = len(unique_classes)
    confusion_matrix = np.zeros((n_classes, n_classes), dtype=int)

    for i in range(len(y_true)):
        true_class = int(y_true[i])
        predicted_class = int(y_pred[i])
        confusion_matrix[true_class, predicted_class] += 1
    
    return confusion_matrix

# Manually calculate accuracy
def calculate_accuracy(y_true, y_pred):
    """
    Computes accuracy manually.
    """
    correct_predictions = sum(y_true[i] == y_pred[i] for i in range(len(y_true)))
    return correct_predictions / len(y_true)

# Calculate and display confusion matrix and accuracy
conf_matrix = calculate_confusion_matrix(y_test, y_pred)
accuracy = calculate_accuracy(y_test, y_pred)

print("\nConfusion Matrix:")
print(conf_matrix)
print(f"\nAccuracy: {accuracy:.2f}")

def calculate_precision_recall_f(conf_matrix):
    """
    Calculates precision, recall, and F1-score for each class and their means.
    """
    precisions = []
    recalls = []
    f_scores = []

    for i in range(len(conf_matrix)):
        true_positive = conf_matrix[i, i]
        false_positive = sum(conf_matrix[:, i]) - true_positive
        false_negative = sum(conf_matrix[i, :]) - true_positive

        # Avoid division by zero
        precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) > 0 else 0
        recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) > 0 else 0
        f_score = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        precisions.append(precision)
        recalls.append(recall)
        f_scores.append(f_score)

    # Mean precision, recall, and F1-score
    mean_precision = np.mean(precisions)
    mean_recall = np.mean(recalls)
    mean_f = np.mean(f_scores)

    return precisions, recalls, f_scores, mean_precision, mean_recall, mean_f

# Call the function with the confusion matrix
precisions, recalls, f_scores, mean_precision, mean_recall, mean_f = calculate_precision_recall_f(conf_matrix)

# Display results
print("\nPrecision for each class:")
for i, precision in enumerate(precisions):
    print(f"Class {i}: {precision:.2f}")

print("\nRecall for each class:")
for i, recall in enumerate(recalls):
    print(f"Class {i}: {recall:.2f}")

print("\nF-score for each class:")
for i, f in enumerate(f_scores):
    print(f"Class {i}: {f:.2f}")

print(f"\nMean Precision: {mean_precision:.2f}")
print(f"Mean Recall: {mean_recall:.2f}")
print(f"Mean F-score: {mean_f:.2f}")
```
